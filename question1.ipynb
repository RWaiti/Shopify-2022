### **Question 1: Given some sample data, write a program to answer the following:**

*On Shopify, we have exactly 100 sneaker shops, and each of these shops sells only one model of shoe. We want to do some analysis of the average order value (AOV). When we look at orders data over a 30 day window, we naively calculate an AOV of $3145.13. Given that we know these shops are selling sneakers, a relatively affordable item, something seems wrong with our analysis.*

```python
import pandas as pd

df = pd.read_csv("2019_Winter_Data_Science_Intern_Challenge_Data_Set_-_Sheet1.csv", index_col=0, encoding="utf-8",sep=',')
df["created_at"] = pd.to_datetime(df["created_at"])
df["week_day"] = df["created_at"].dt.day_name()
```

```python
total_amount = float(sum(df['order_amount']))
total_items = sum(df['total_items'])
total_sell = len(df)
```

## a) Think about what could be going wrong with our calculation. Think about a better way to evaluate this data.

#### What we can see is that the AOV was calculated using the total amount of sales, not the total amount of sold shoes.


```python
print("Total sells:", total_sell)
print("total amount: ${:.2f}".format(total_amount))
print("Wrong AOV: ${:.2f}".format(total_amount/total_sell))
```
